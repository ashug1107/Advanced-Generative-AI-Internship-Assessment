```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import textwrap

def wrap_text(text, width=80):
    """Wrap text to fit screen width"""
    return textwrap.fill(text, width=width, subsequent_indent='    ')

tokenizer = AutoTokenizer.from_pretrained("Locutusque/gpt2-medium-conversational")
model = AutoModelForCausalLM.from_pretrained("Locutusque/gpt2-medium-conversational")
tokenizer.pad_token = tokenizer.eos_token

print("=" * 60)
print("🚀 TRANSFORMERS CHATBOT - NLP Assignment 3")
print("=" * 60)
print("Chatbot: Hello! I am your AI assistant. How can I help you today?")
print("-" * 60)

# Simple: No full history, regenerate per turn with system prompt
system_prompt = "You are a helpful assistant. Give short, accurate answers."

while True:
    user_input = input("User: ").strip()
    print(f"👤 Human: {user_input}")
    
    if user_input.lower() in ['exit', 'quit']:
        print("🤖 Chatbot: Goodbye! 👋")
        print("=" * 60)
        print("✅ Chatbot session ended successfully!")
        print("=" * 60)
        break
    
    # Concise prompt
    prompt = f"{system_prompt}\nUser: {user_input}\nAssistant:"
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    
    # Tight generation: SHORT, controlled
    outputs = model.generate(
        input_ids,
        max_new_tokens=40,      # Short replies
        temperature=0.3,        # Low randomness = factual
        top_p=0.85,
        repetition_penalty=1.2, # Heavy anti-repeat
        no_repeat_ngram_size=2,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    
    response = tokenizer.decode(outputs[0][input_ids.shape[1]:], skip_special_tokens=True).strip()
    wrapped_response = wrap_text(response)
    print(f"🤖 Chatbot: {wrapped_response}")
    print("-" * 60)
```
