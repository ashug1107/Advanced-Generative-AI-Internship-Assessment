```python
#=================================================================
# 2. IMPORTS (warnings FIXED)
# =================================================================
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import re
import warnings  # ✅ FIXED: Added missing import
warnings.filterwarnings('ignore')

print("✅ All imports successful!")

# =================================================================
# 3. LOAD & PREPROCESS Twitter_Data.csv
# =================================================================
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Twitter_Data.csv')
print(f"✅ Loaded: {df.shape}")
print("Columns:", df.columns.tolist())
print(df.head())

# Select columns & fix labels
df = df[['clean_text', 'category']].copy()
print("\nRaw category distribution:")
print(df['category'].value_counts(dropna=False))

# ✅ FIX: Remove invalid labels (-1, NaN)
df = df.dropna(subset=['category'])
df = df[df['category'] >= 0]  # Remove -1

# ✅ Binary classification (0=negative, 1=positive/neutral)
df['label'] = (df['category'] > 0).astype(int)
print("\n✅ Fixed labels (0/1 only):")
print(df['label'].value_counts())

# Twitter text cleaning
def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+|@\w+|#\w+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['clean_text'] = df['clean_text'].fillna('').apply(clean_tweet)
df = df[df['clean_text'].str.len() > 5]  # Remove empty/short tweets
print(f"\n✅ Final clean dataset: {df.shape}")

# =================================================================
# 4. TRAIN/VAL/TEST SPLIT
# =================================================================
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df['label'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['label'])
print(f"✅ Train: {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")

# =================================================================
# 5. FIXED TRANSFER LEARNING DATASETS (KeyError 52456 FIXED)
# =================================================================
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
print("✅ Pre-trained BERT tokenizer loaded!")

class TwitterDataset(Dataset):
    def __init__(self, df_texts, df_labels, tokenizer, max_len=128):
        # 🔥 KEY FIX: Reset index + convert to numpy arrays
        self.texts = df_texts.reset_index(drop=True).values
        self.labels = df_labels.reset_index(drop=True).values.astype(int)
        self.tokenizer = tokenizer
        self.max_len = max_len
        print(f"✅ Dataset created: {len(self.texts)} samples")

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        # 🔥 SAFE ACCESS: Continuous index guaranteed
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        encoding = self.tokenizer(
            text,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# ✅ FIXED: Pass reset_index Series
train_dataset = TwitterDataset(train_df['clean_text'], train_df['label'], tokenizer)
val_dataset = TwitterDataset(val_df['clean_text'], val_df['label'], tokenizer)
test_dataset = TwitterDataset(test_df['clean_text'], test_df['label'], tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=16, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=16, num_workers=0)

print("✅ FIXED DATASETS READY!")

# =================================================================
# 6. DEVICE & TRAINING FUNCTIONS
# =================================================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Device: {device}")

def train_epoch(model, data_loader, optimizer, scheduler=None):
    model.train()
    total_loss = 0
    loop = tqdm(data_loader, desc="Training")

    for batch in loop:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        # TRANSFER LEARNING forward pass
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        if scheduler:
            scheduler.step()

        total_loss += loss.item()
        loop.set_postfix({'loss': f'{loss.item():.4f}'})

    return total_loss / len(data_loader)

def evaluate(model, data_loader):
    model.eval()
    predictions = []
    true_labels = []

    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask)
            _, preds = torch.max(outputs.logits, dim=1)
            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    metrics = compute_metrics(true_labels, predictions)
    return metrics, predictions, true_labels

def compute_metrics(true_labels, predictions):
    accuracy = accuracy_score(true_labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, predictions, average='weighted')
    return {'accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1': f1}

# =================================================================
# 7. EXPERIMENT 1: FULL TRANSFER LEARNING
# =================================================================
print("\n" + "="*70)
print("EXPERIMENT 1: FULL TRANSFER LEARNING (All layers)")
print("="*70)

# ✅ PRE-TRAINED BERT TRANSFER LEARNING
model1 = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2).to(device)
optimizer1 = AdamW(model1.parameters(), lr=2e-5)
scheduler1 = get_linear_schedule_with_warmup(optimizer1, 0, len(train_loader)*2)

print("🧠 Training Full Transfer Learning...")
for epoch in range(2):
    print(f"\nEpoch {epoch+1}/2")
    train_epoch(model1, train_loader, optimizer1, scheduler1)

metrics1, preds1, trues1 = evaluate(model1, test_loader)
print(f"✅ Full TL - Accuracy: {metrics1['accuracy']:.4f}, F1: {metrics1['f1']:.4f}")

# =================================================================
# 8. EXPERIMENT 2: FREEZE BERT BASE
# =================================================================
print("\n" + "="*70)
print("EXPERIMENT 2: FREEZE BERT (Classifier only)")
print("="*70)

model2 = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2).to(device)
for param in model2.bert.parameters():
    param.requires_grad = False  # Freeze pre-trained BERT

optimizer2 = AdamW(model2.parameters(), lr=2e-5)
print("⚡ Training Frozen BERT + Classifier...")
for epoch in range(2):
    train_epoch(model2, train_loader, optimizer2)

metrics2, preds2, trues2 = evaluate(model2, test_loader)
print(f"✅ Frozen BERT - Accuracy: {metrics2['accuracy']:.4f}, F1: {metrics2['f1']:.4f}")

# =================================================================
# 9. EXPERIMENT 3: LAST 2 LAYERS
# =================================================================
print("\n" + "="*70)
print("EXPERIMENT 3: FINE-TUNE LAST 2 BERT LAYERS")
print("="*70)

model3 = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2).to(device)
for name, param in model3.named_parameters():
    if 'layer.10' not in name and 'layer.11' not in name:
        param.requires_grad = False  # Freeze all but last 2 layers

optimizer3 = AdamW(model3.parameters(), lr=2e-5)
print("🎯 Training Last 2 Layers...")
for epoch in range(2):
    train_epoch(model3, train_loader, optimizer3)

metrics3, preds3, trues3 = evaluate(model3, test_loader)
print(f"✅ Last 2 Layers - Accuracy: {metrics3['accuracy']:.4f}, F1: {metrics3['f1']:.4f}")

# =================================================================
# 10. RESULTS TABLE & PLOTS
# =================================================================
results_df = pd.DataFrame({
    'Experiment': ['Full Transfer Learning', 'Frozen BERT', 'Last 2 Layers'],
    'Accuracy': [metrics1['accuracy'], metrics2['accuracy'], metrics3['accuracy']],
    'Precision': [metrics1['precision'], metrics2['precision'], metrics3['precision']],
    'Recall': [metrics1['recall'], metrics2['recall'], metrics3['recall']],
    'F1-Score': [metrics1['f1'], metrics2['f1'], metrics3['f1']]
}).round(4)

print("\n" + "="*80)
print("🏆 FINAL TRANSFER LEARNING RESULTS")
print("="*80)
print(results_df.to_string(index=False))

# Visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Bar comparison
width = 0.3
x = np.arange(3)
axes[0,0].bar(x-width/2, results_df['Accuracy'], width, label='Accuracy', alpha=0.8)
axes[0,0].bar(x+width/2, results_df['F1-Score'], width, label='F1-Score', alpha=0.8)
axes[0,0].set_title('Transfer Learning Comparison')
axes[0,0].set_xticks(x)
axes[0,0].set_xticklabels(results_df['Experiment'], rotation=20)
axes[0,0].legend()

# Confusion Matrix (Best: Full Transfer Learning)
cm = confusion_matrix(trues1, preds1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0,1])
axes[0,1].set_title('Confusion Matrix - Full Transfer Learning')

# Data distribution
df['label'].value_counts().plot.pie(ax=axes[1,0], autopct='%1.1f%%')
axes[1,0].set_title('Twitter Sentiment Distribution')

# Sample predictions
sample_texts = test_df['clean_text'].head(5).values
sample_preds = preds1[:5]
sample_trues = trues1[:5]
table_data = [f"Tweet: {t[:50]}...\nPred: {p}, True: {tr}"
              for t, p, tr in zip(sample_texts, sample_preds, sample_trues)]
axes[1,1].axis('off')
axes[1,1].text(0.05, 0.95, '\n\n'.join(table_data), va='top', fontsize=10)
axes[1,1].set_title('Sample Predictions')

plt.tight_layout()
plt.show()
```
