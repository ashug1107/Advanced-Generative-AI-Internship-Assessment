1. User Login Check

```python
# Hardcoded credentials
VALID_USERNAME = "admin"
VALID_PASSWORD = "1234"

# Get user input
username = input("Enter username: ")
password = input("Enter password: ")

# Authentication logic
if username == VALID_USERNAME and password == VALID_PASSWORD:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
# Student marks list
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final results
print("--- Academic Evaluation Summary ---")
print(f"Total Students: {len(marks)}")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
```

3. Simple Data Cleaner

```python
# Raw data
names = [" Alice ", "bob", " CHARLIE "]

# Data Cleaning: Strip spaces, convert to lowercase and convert 1st letter into Capital
cleaned_names = [name.strip().lower().title() for name in names]

# Print results
print("Original List:", names)
print("Cleaned List: ", cleaned_names)
```

4. Message Length Analyzer

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

print(f"{'Message':<25} | {'Length':<6} | {'Status'}")
print("-" * 45)

# Process each message
for msg in messages:
    length = len(msg)
    # Flag messages longer than 10 characters
    status = "LONG" if length > 10 else "NORMAL"
    
    print(f"{msg:<25} | {length:<6} | {status}")
```

5. Error Message Detector

```python
# System log data
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# A robust way to count regardless of casing
robust_count = sum(1 for log in logs if log.upper() == "ERROR")
print(f"Robust Error Count: {robust_count}")
```
