```python
!pip install ctransformers langchain-community

from langchain_community.llms import CTransformers
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Initialize a tiny local model 
# This runs on your CPU, so no API keys or 403 errors are possible.
llm = CTransformers(
    model="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
    model_file="tinyllama-1.1b-chat-v1.0.Q4_K_M.gguf",
    model_type="llama",
    config={'max_new_tokens': 256, 'temperature': 0.7}
)

# 2. Define your Prompt
template = """<|system|>
You are a creative storyteller.</s>
<|user|>
Explain {topic} in a {style} style.</s>
<|assistant|>"""

prompt = PromptTemplate.from_template(template)

# 3. Build the Chain
chain = prompt | llm | StrOutputParser()

# 4. Run it
try:
    print("--- Running Locally on your CPU (No API Key needed) ---")
    result = chain.invoke({"topic": "Neural Networks", "style": "storytelling"})
    print(result)
except Exception as e:
    print(f"Local Error: {e}")
```
