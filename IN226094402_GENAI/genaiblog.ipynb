```python
!pip install ctransformers langchain-community

import textwrap
from langchain_community.llms import CTransformers
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import BaseOutputParser

# 1. Custom Parser to handle line wrapping
class WrappedOutputParser(BaseOutputParser[str]):
    """Custom parser to wrap text at a specific width."""
    width: int = 80  # Set your desired character limit per line

    def parse(self, text: str) -> str:
        # Wraps the text so it fits the screen width
        return "\n".join(textwrap.wrap(text, width=self.width))

# 2. Initialize Local Model (No API needed)
llm = CTransformers(
    model="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
    model_file="tinyllama-1.1b-chat-v1.0.Q4_K_M.gguf",
    model_type="llama",
    config={'max_new_tokens': 300, 'temperature': 0.7}
)

# 3. Prompt Template
template = """<|system|>
You are a creative storyteller.</s>
<|user|>
Explain {topic} in a {style} style.</s>
<|assistant|>"""

prompt = PromptTemplate.from_template(template)

# 4. The Chain (Prompt -> LLM -> Custom Wrapped Parser)
chain = prompt | llm | WrappedOutputParser(width=70)

# 5. Run
try:
    print(f"{'='*70}\nRunning Locally: Output will wrap at 70 characters\n{'='*70}")
    result = chain.invoke({"topic": "Neural Networks", "style": "storytelling"})
    print(result)
except Exception as e:
    print(f"Error: {e}")
```
