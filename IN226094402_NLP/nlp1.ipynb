```python
#Task 1: Conceptual Understanding (Mandatory – Written)

#1. What is the difference between "Love" and "love" in NLP?
'''For a computer, these two are completely different.
1.The ASCII/Binary Factor: Computers process text based on character codes. "L" (uppercase) and "l" (lowercase) have different numeric values, so "Love" and "love" are treated as two unique tokens.
2.The NLP Impact: If you don't perform Lowercasing (a common preprocessing step), your model will count them separately. This splits your data—for example, a sentiment analysis tool might learn that "love" is positive but wouldn't know what "Love" means because it treats it as a brand-new word.'''

#2. What happens if stopwords are not removed?
'''Stopwords are high-frequency words like "the," "is," and "and." If you leave them in:
1.Noise & Clutter: They act as "noise." Because they appear everywhere, they don't carry much unique meaning. The model might think "the" is the most important word in a document just because it appears 50 times.
2.Increased Dimensionality: Your dataset becomes much larger and heavier. This slows down training and requires more memory/computation power for no real benefit.
3.Inaccuracy: In tasks like Document Classification, the model might fail to see the "signal" (the important keywords) because it's buried under a mountain of common filler words.'''

#3. Provide two real-world scenarios where removing stopwords can be harmful.
'''Sometimes, these "filler" words are actually the most important part of the sentence:
Sentiment Analysis (Negation): In the sentence "I do not like this movie," the word "not" is often considered a stopword. If you remove it, the sentence becomes "I like this movie." Your model will predict a positive sentiment when the user actually meant the opposite.
Search Queries for Specific Titles: If you are searching for the movie "The It" or the band "The The", a system that removes stopwords will strip away the entire title, leaving the search engine with nothing to look for.'''

#4. Explain the difference between stemming and lemmatization.
'''Both are used to reduce a word to its "base" form, but they do it differently:
Stemming: Think of this as a "Chop" method. It follows a set of crude rules to hack off the ends of words (like "-ing," "-ed," or "-ies").
Example: "Caring" might become "Car." It’s fast, but it often results in "stems" that aren't real dictionary words.
Lemmatization: Think of this as the "Smart" method. It looks at the context and the dictionary (Morphological analysis). It understands the part of speech (is it a verb or a noun?).
Example: It knows that "better" is a form of the word "good." It will return a real, meaningful root word (called a Lemma). It is slower and more complex but much more accurate.'''
```

```python
#Task 2:Build Advanced Preprocessing Function

import re
import json
from collections import Counter

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs and Email-like patterns
    text = re.sub(r'https?://\S+|www\.\S+|<\S+@\S+>', '', text)
    
    # 3. Handle repeated characters
    text = re.sub(r'(.)\1+', r'\1\1', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Remove punctuation/symbols
    text = re.sub(r'[^\w\s]', '', text)
    
    # 6. Remove very short tokens (Length <= 2, keep "no", "not")
    keep_words = {'no', 'not'}
    words = text.split()
    filtered_words = [w for w in words if len(w) > 2 or w in keep_words]
    
    return filtered_words, " ".join(filtered_words)

# --- Testing the implementation ---
test_string = "Visit http://example.com now! I have 2 dogs. It is soooo goooood!!! WOW!!! This is GREAT!!! no not bad"
print(preprocess_text(test_string))
```

```python
#Task 3: Stress Testing

# Sample Inputs
inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Execution and Formatting
print(f"{'ORIGINAL TEXT':<40} | {'CLEANED TOKENS':<30} | {'CLEANED SENTENCE'}")
print("-" * 110)

for item in inputs:
    tokens, sentence = preprocess_text(item)
    print(f"{item:<40} | {str(tokens):<30} | {sentence}")
```

```python
#Task 5: Frequency Analysis

all_tokens = []

for item in inputs:
    tokens, sentence = preprocess_text(item)
    all_tokens.extend(tokens)

# 2. Use Counter for frequency analysis
word_counts = Counter(all_tokens)

# 3. Identify Top 10 most frequent
top_10 = word_counts.most_common(10)

# 4. Identify Top 5 least frequent
least_5 = word_counts.most_common()[:-6:-1]

print("--- TOP 10 MOST FREQUENT WORDS ---")
for word, count in top_10:
    print(f"{word}: {count}")

print("\n--- TOP 5 LEAST FREQUENT WORDS ---")
for word, count in least_5:
    print(f"{word}: {count}")
```

```python
#Task 6: Build Full Pipeline

def full_pipeline(text_list):
    """
    Processes a list of strings through the preprocessing engine.
    Returns: A dictionary with aggregated tokens and cleaned sentences.
    """
    # Initialize the required output structure
    pipeline_result = {
        "tokens": [],
        "clean_sentences": []
    }
    
    for text in text_list:
        # Call the preprocessing function for each item
        tokens, sentence = preprocess_text(text)
        
        # Aggregate tokens into one flat list
        pipeline_result["tokens"].extend(tokens)
        
        # Add the cleaned sentence to the sentences list
        pipeline_result["clean_sentences"].append(sentence)
        
    return pipeline_result

# --- Implementation / Testing ---

# Sample Inputs from the assignment
sample_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "I am not happy with this"
]

# Run the pipeline
final_output = full_pipeline(sample_data)

# Print the result in a readable JSON format
print(json.dumps(final_output, indent=4))
```
