```python
#Problem 1 : Smart Light Controller

light_name = "Bedroom Light"
action = "ON"

class Smart_Light:
    def __init__(self, light_name, action):
        self.light_name = light_name
        self.action = action

    def light_control(self):
        if(self.action=="ON"):
            print(f'{self.light_name} is ON')
        else:
            print(f'{self.light_name} is OFF')

light1 = Smart_Light(light_name, action)
light1.light_control()
```

```python
#Problem 2: Employee ID Card System

Employee_Name = "Rahul"
Employee_ID = "EMP102"
Department = "IT"

class Employee_Details:
    def __init__(self, name, id, department):
        self.name = name
        self.id = id
        self.department = department
    
    def display(self):
        print(f"Employee ID Card")
        print(f"Name : {self.name}")
        print(f'ID : {self.id}')
        print(f"Department : {self.department}")

employee = Employee_Details(Employee_Name, Employee_ID, Department)
employee.display()
```

```python
#Problem 3: Mobile Contact Record

Contact_Name = "Anita"
Phone_Number = 9876543210

class mobile_contact:
    def __init__(self, name, number):
        self.name = name
        self.number = number

    def save_contact(self):
        print("Contact Saved")
        print(f"Name : {self.name}")
        print(f"Phone : {self.number}")

contact = mobile_contact(Contact_Name, Phone_Number)
contact.save_contact()
```

```python
#Problem 4: Product Price Tag Generator

Product_Name = "Headphones"
Price = 2499

class product_price_tag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def generate_price_tag(self):
        print(f'Product: {self.name}')
        print(f'Price: {self.price}')

product = product_price_tag(Product_Name, Price)
product.generate_price_tag()
```

```python
#Problem 5: Movie Rating Display System

Movie_Name = "Inception"
Rating = 4.8

class movie_rating_system:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print(f'Movie : {self.name}')
        print(f'Rating : {self.rating}')
    
movie = movie_rating_system(Movie_Name, Rating)
movie.display()
```
