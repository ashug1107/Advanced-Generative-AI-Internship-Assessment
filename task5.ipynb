```python
#Problem 1 : Social Media - Post Engagement Analyzer

def analyze_engagement(likes_list):
    # 1. Initialize total likes counter
    total_likes = 0
    
    # 2. Use a loop to calculate total likes from different sources
    for likes in likes_list:
        total_likes += likes
        
    # 3. Determine post status based on the 1000 likes threshold
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
        
    return total_likes, status

# Input: Likes from Organic, Paid, and Shared sources
post_engagement = [450, 300, 300]

# Running the analysis
total, post_status = analyze_engagement(post_engagement)

# Displaying the expected output
print(f"Total Likes: {total}")
print(f"Post Status: {post_status}")
```

```python
#Problem 3 : Agriculture – Rainfall Adequacy Checker

def analyze_seasonal_rainfall(rainfall_data, required_level=70):
    # 1. Initialize total rainfall sum
    total_rainfall = 0
    
    # 2. Use a loop to calculate the sum of all seasonal records
    for record in rainfall_data:
        total_rainfall += record
        
    # 3. Calculate average rainfall
    # Using len() to find the total number of records
    count = len(rainfall_data)
    average_rainfall = total_rainfall / count if count > 0 else 0
        
    # 4. Determine status based on the required level
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
        
    return average_rainfall, status

# Input: Rainfall recorded across different months/weeks
seasonal_data = [65, 80, 72, 68, 75]

# Running the analysis
avg, rainfall_status = analyze_seasonal_rainfall(seasonal_data)

# Displaying the expected output
print(f"Average Rainfall: {int(avg)}")
print(f"Rainfall Status: {rainfall_status}")
```

```python
#Problem 4: Social Media - Duplicate Account Detection

def process_usernames(new_usernames):
    seen_usernames = set()
    results = {}

    for username in new_usernames:
        # Check if username is already in our 'seen' collection
        if username in seen_usernames:
            results[username] = "Flagged: Duplicate Account"
        else:
            results[username] = "Unique: Account Verified"
            # Add the new unique username to the set
            seen_usernames.add(username)
            
    return results

# Input: A list of usernames signing up for the system
users_to_check = ["alice_99", "bob_dev", "alice_99", "charlie_k"]

# Running the detection
verification_report = process_usernames(users_to_check)

# Displaying the results
for user, status in verification_report.items():
    print(f"{user}: {status}")
```

```python
#Problem 6 : Agriculture – Premium Crop Price Filter

def filter_premium_crops(price_list):
    # 1. Initialize an empty list to store premium prices
    premium_crops = []
    
    # 2. Use a loop to check each price against the threshold
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
            
    return premium_crops

# Input: Current market prices for various crop batches (in ₹)
crop_prices = [1500, 2500, 1800, 3200, 2000, 1200]

# Running the filter
premium_list = filter_premium_crops(crop_prices)

# Displaying the expected output
print(f"Premium Crops: {premium_list}")
```

```python
#Problem 7 : System Monitoring – Application Health Checker

def check_app_health(error_count):
    # 1. Check for the most stable state
    if error_count == 0:
        status = "Healthy"
    # 2. Check for the intermediate threshold
    elif error_count <= 5:
        status = "Minor Issues"
    # 3. Handle everything above the threshold
    else:
        status = "Critical Issues"
        
    return status

# Input: The number of errors detected in the last monitoring cycle
current_errors = 7

# Running the health check
system_status = check_app_health(current_errors)

# Displaying the expected output
print(f"Error Count: {current_errors}")
print(f"System Status: {system_status}")
```

```python
#Problem 8 : Banking – Daily Transaction Limit Checker

def check_transaction_limit(amount):
    # 1. Define the daily limit constant
    DAILY_LIMIT = 50000
    
    # 2. Compare the requested amount against the limit
    if amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"
        
    return status

# Input: The amount the user is trying to transfer
transaction_amount = 60000

# Running the validation check
transaction_status = check_transaction_limit(transaction_amount)

# Displaying the expected output
print(f"Transaction Amount: {transaction_amount}")
print(f"Transaction Status: {transaction_status}")
```

```python
#Problem 9 : E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    # 1. Count 'P' (Present) using a loop or count method
    present_count = 0
    for status in attendance_list:
        if status == "P":
            present_count += 1
            
    # 2. Get total number of sessions
    total_sessions = len(attendance_list)
    
    # 3. Calculate percentage
    # Formula: (Present / Total) * 100
    if total_sessions > 0:
        attendance_percentage = (present_count / total_sessions) * 100
    else:
        attendance_percentage = 0.0
        
    # 4. Apply the eligibility rule
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
        
    return attendance_percentage, eligibility

# Input: A list of attendance records
attendance_data = ["P", "P", "P", "A", "P"]

# Running the system
percent, status = check_exam_eligibility(attendance_data)

# Displaying the expected output
print(f"Attendance Percentage: {percent:.1f}")
print(f"Exam Eligibility: {status}")
```

```python
#Problem 10 : Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    total_bill = 0
    
    # 1. Billing Logic - Slab Calculation
    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        # First 100 at ₹3, remaining at ₹5
        total_bill = (100 * 3) + ((units - 100) * 5)
    else:
        # First 100 at ₹3, next 100 at ₹5, rest at ₹7
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
        
    # 2. Usage Classification
    if total_bill < 500:
        usage_status = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
        
    return total_bill, usage_status

# Input: Units consumed by the user
units_consumed = 250

# Running the calculation
bill_amount, status = calculate_electricity_bill(units_consumed)

# Displaying the expected output
print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Status: {status}")
```
