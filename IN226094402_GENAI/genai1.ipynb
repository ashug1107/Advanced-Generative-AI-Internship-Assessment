```python
from langchain_core.prompts import PromptTemplate

# Define the template
template = "Explain {topic} in simple terms for beginners"

# Initialize PromptTemplate
prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Reusable function
def explain_topic_langchain(topic_name):
    return prompt_template.format(topic=topic_name)

print(explain_topic_langchain("Quantum Computing"))
```

```python
multi_input_template = "Explain {topic} for {audience} in a {tone} tone"

prompt_multi = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

# Test Cases
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for test in test_cases:
    print(prompt_multi.format(**test))
```

```python
teaching_prompt = PromptTemplate.from_template("Teaching → Explain {topic} clearly step by step")
interview_prompt = PromptTemplate.from_template("Interview → Ask 3 questions about {topic}")
story_prompt = PromptTemplate.from_template("Storytelling → Explain {topic} as a story")

topic = "Machine Learning"
print(teaching_prompt.format(topic=topic))
print(interview_prompt.format(topic=topic))
print(story_prompt.format(topic=topic))
```

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

def build_chat_prompt(role, topic):
    system_template = "You are a helpful {role}."
    human_template = "Explain {topic}."
    
    chat_prompt = ChatPromptTemplate.from_messages([
        SystemMessagePromptTemplate.from_template(system_template),
        HumanMessagePromptTemplate.from_template(human_template),
    ])
    
    return chat_prompt.format_prompt(role=role, topic=topic).to_messages()

# Example
print(build_chat_prompt(role="teacher", topic="Neural Networks"))
```

```python
def validate_inputs(audience, tone):
    valid_audiences = ['beginner', 'intermediate', 'expert']
    valid_tones = ['formal', 'casual', 'fun']
    
    # Validation Logic
    final_audience = audience if audience in valid_audiences else 'beginner'
    final_tone = tone if tone in valid_tones else 'formal'
    
    return final_audience, final_tone

# Testing validation
a, t = validate_inputs("newbie", "excited") # Invalid inputs
print(f"Validated values: Audience={a}, Tone={t}")
```

```python
def generate_prompt(topic, audience, tone, style):
    # 1. Validation
    audience, tone = validate_inputs(audience, tone)
    
    # 2. Styles mapping
    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask 3 questions about {topic} for an {audience} in a {tone} tone",
        "storytelling": "Tell a {tone} story about {topic} for {audience}"
    }
    
    # 3. Dynamic Template Selection
    template = styles.get(style, styles["teaching"])
    prompt_obj = PromptTemplate(input_variables=["topic", "audience", "tone"], template=template)
    
    return prompt_obj.format(topic=topic, audience=audience, tone=tone)

# Task 7: Reusability Test
inputs = ["Neural Networks", "FastAPI", "Docker", "SQL", "Cloud Computing"]
for item in inputs:
    print(generate_prompt(item, "beginner", "fun", "storytelling"))
```
