```python
#Problem 1 : Smart Parking Lot Management System

Parking_Capacity = 50
Vehicle_Logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

def parking_management(Vehicle_Logs, Parking_Capacity):
    pcount = 0    #parking count
    ocount = 0    #exited cars
    for parked in Vehicle_Logs:
        if (parked=="IN"):
            pcount = pcount+1
        else:
            ocount = ocount+1
    total_parked = pcount - ocount
    print(f"Total Cars Parked in Parking Lot : {total_parked}")
    if(total_parked<50):
        print(f"Parking Status: Available")
    else:
        print("Alert!! Parking Limit Exceeded. Parking Status : Unavailable")

parking_management(Vehicle_Logs, Parking_Capacity)
```

```python
#Problem 3 : Movie Theatre Seat Occupancy Analyzer

def analyze_seat_booking(total_seats, booked_seats_list):
    num_booked = len(booked_seats_list)
    
    occupancy = (num_booked / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Half Full"
    else:
        status = "Low Occupancy"
        
    return occupancy, status

total_capacity = 200
booked_data = [1] * 150 

occ_percent, show_status = analyze_seat_booking(total_capacity, booked_data)

print(f"Occupancy: {occ_percent}%")
print(f"Show Status: {show_status}")
```

```python
#Problem 4 : Cloud Server Load Classification System

CPU_Readings = [45, 60, 70, 85, 90]
count = len(CPU_Readings)

def avg_cpu_load(CPU_Readings, count):
    total_readings = 0
    for readings in CPU_Readings:
        total_readings += readings
    avg_load = total_readings/count
    print(f'Average CPU Load : {avg_load}')
    if (avg_load<50):
        print('Server Status : Normal')
    elif(avg_load>=50 and avg_load<=80):
        print("Server Status : Warning")
    else:
        print('Server Status : Critical')

avg_cpu_load(CPU_Readings, count)
```

```python
#Problem 5 : Smart Classroom Resource Usage Monitor

Resource_Usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

def resource_usage(Resource_Usage):
    highest_resource = max(Resource_Usage, key=Resource_Usage.get)
    highest_value = Resource_Usage[highest_resource]
    print(f'Overused Resources : {highest_resource}')
    if (highest_value>5):
        print('Energy Alert: Yes')

resource_usage(Resource_Usage)
```

```python
#Problem 6 : Online Event Registration Capacity Controller

Event_Capacity = 100
Registrations = 105

def event_registration(Event_Capacity, Registrations):
    if(Event_Capacity<Registrations):
        waitlist = Registrations - Event_Capacity
        print(f"Confirmed Registrations : {Event_Capacity}")
        print(f"Waitlisted Users : {waitlist}")
        print(f"Registration Status : Closed")
    elif(Event_Capacity>Registrations):
        print(f"Confirmed Registrations : {Registrations}")
        print(f"Registration Status : Open")

event_registration(Event_Capacity, Registrations)
```
