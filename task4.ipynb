```python
#Problem 1 : Mobile Recharge Validation System

def mobile_recharge_validation():
    print('Recharge Plans : ')
    print('1. 199\n2. 299\n3. 399\n4. 599')
    recharge = int(input('Your Selected Recharge Plan : '))
    if recharge==199 or recharge==299 or recharge==399 or recharge==599 and recharge>=50:
        print('Recharge Plan Succesful!!!')
    else:
        print('Recharge Plan is Invalid!! Try Again.')

n = 0
while(n<5):
    mobile_recharge_validation()
    n = n + 1
```

```python
#Problem 2 : Inventory Reorder Alert System 

dict = {'Table' : 10, 'Chair' : 15, 'Sofa' : 12, 'Book Shelf' : 30, 'Dining Table' : 14}

def reorder_alert(dict):
    for key, value in dict.items():
        if value<15 :
            print(f'Reorder Alert : Reorder {key}, Stock Available {value}')
        else:
            print(f'Stock OK for {key}!! Stock is {value}')

reorder_alert(dict)
```

```python
#Problem 3 : Student Result Processing System

student_marks = []

def student_mark():
    print('Enter Marks of 5 subjects of a Student\n')
    n = 0
    total = 0
    while(n<5):
        marks = int(input(f'Enter Marks of Subject {n+1} : '))
        student_marks.append(marks)
        n = n + 1
        total += marks
    avg_marks = total/5
    print(f'Student List of Marks : {student_marks}')
    print(f'Average Marks of Student : {avg_marks}')
    if(avg_marks<=50):
        print('Result : Fail')
    else:
        print('Result : Pass')

student_mark()
```

```python
#Problem 4 : Cab Fare Estimator with Retry Option

base_fare = 50

def cab_fare_calc():
    kilometer = int(input('Enter Kilometer to Travel : '))
    print(f'Kilometers to Travel : {kilometer}')
    final_fare = kilometer*12 + base_fare
    is_it_peak_hour = input('Enter "Yes" or "No" if it is Peak Hour : ')
    if(is_it_peak_hour=="Yes"):
        peakhour_fare = final_fare*25/100
        final_fare += peakhour_fare
        print(f'Final Fare at Peak Hour : {final_fare}')
    else:
        print(f'Final Estimated Fare : {final_fare}')
    print('Do you want Retry : \nWrite "Yes" or "No"\n')
    retry = input('Enter Choice : ')
    if(retry=="Yes"):
        cab_fare_calc()
    else:
        exit

cab_fare_calc()
```

```python
#Problem 5 : Employee Attendance Eligibility Checker

def check_attendance_eligibility(attendance_list):
    
    present_days = attendance_list.count("P")
    
    
    total_days = len(attendance_list)
    
    if total_days == 0:
        return "Invalid Data"
        
    attendance_percentage = (present_days / total_days) * 100
    
    if attendance_percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"
        
    return result, attendance_percentage

attendance = ["P", "P", "A", "P", "P", "A", "P"]

status, percentage = check_attendance_eligibility(attendance)

print(f"Total Days: {len(attendance)}")
print(f"Attendance Percentage: {percentage:.2f}%")
print(f"Final Status: {status}")
```

```python
#Problem 6 : Password Strength Checker

def check_password_strength(password):
    length_ok = len(password) >= 8
    
    has_digit = any(char.isdigit() for char in password)
    
    special_chars = "@#$"
    has_special = any(char in special_chars for char in password)
    
    if length_ok and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Ensure 8+ chars, 1 digit, and 1 special char @#$)"

print(check_password_strength("Pass123"))      # Weak (too short)
print(check_password_strength("Password@12"))  # Strong
```
