```python
#1. Unique words in sentence

# 1. Define the sentence
sentence = "python is easy and python is powerful"

# 2. Split sentence into words (creates a list)
words = sentence.split()

# 3. Convert list into a set (automatically removes duplicates)
unique_words = set(words)

# 4. Display results
print(f"Original Sentence: {sentence}")
print(f"Unique Words: {unique_words}")
print(f"Total Unique Count: {len(unique_words)}")
```

```python
#2. Highest Salary from employee data

# 1. Store data in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# 2. Find the key (name) associated with the maximum value (salary)
highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

# 3. Display the result
print(f"Employee Records: {employees}")
print("-" * 30)
print(f"Highest Salaried Employee: {highest_paid_employee}")
print(f"Salary Amount: {highest_salary}")
```

```python
#3. Find Minimum and Maximum 

# 1. Input list
numbers = [45, 22, 89, 10, 66]

# 2. Initialize max and min with the first element
max_val = numbers[0]
min_val = numbers[0]

# 3. Use a loop to compare each number
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

# 4. Display results
print(f"List of numbers: {numbers}")
print(f"Maximum Value: {max_val}")
print(f"Minimum Value: {min_val}")
```

```python
#4. Count Products Above a Price Threshold

# 1. Input list of product prices
prices = [450, 1200, 899, 1500, 300]

# 2. Initialize the counter
count_above_1000 = 0

# 3. Use a loop to compare each price
for price in prices:
    if price > 1000:
        count_above_1000 += 1

# 4. Display the result
print(f"Product Prices: {prices}")
print("-" * 30)
print(f"Total products above 1000: {count_above_1000}")
```

```python
#5. Calculate Attendance Percentage

# 1. Input list of attendance
attendance = ["P", "P", "A", "P", "P"]

# 2. Count the number of 'P' (Present) entries
present_count = attendance.count("P")

# 3. Calculate total number of days
total_days = len(attendance)

# 4. Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# 5. Display the results
print(f"Attendance Records: {attendance}")
print("-" * 35)
print(f"Total Days Tracked: {total_days}")
print(f"Total Days Present: {present_count}")
print(f"Attendance Percentage: {attendance_percentage}%")
```

```python
#6. Remove Duplicate Phone Numbers

# 1. Input list with duplicate phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# 2. Convert the list into a set to automatically remove duplicates
unique_contacts = set(phone_numbers)

# 3. Display results
print(f"Original List: {phone_numbers}")
print(f"Unique Phone Numbers: {unique_contacts}")
```

```python
#7. Count Character Frequency

# 1. Input string
text = "python"

# 2. Initialize an empty dictionary to store counts
char_freq = {}

# 3. Iterate through each character
for char in text:
    char_freq[char] = char_freq.get(char, 0) + 1

# 4. Display results
print(f"Original String: '{text}'")
print("-" * 30)
print("Character Frequencies:")
for char, count in char_freq.items():
    print(f"'{char}': {count}")
```

```python
#8. Convert List to Tuple

# 1. Input list
data_list = [10, 20, 30]

# 2. Convert to tuple using the tuple() function
data_tuple = tuple(data_list)

# 3. Display results and data types
print(f"Original List: {data_list} | Type: {type(data_list)}")
print(f"Converted Tuple: {data_tuple} | Type: {type(data_tuple)}")
```

```python
#9. Check if a Key Exists in a Dictionary

# 1. Define a dictionary
user_data = {
    "id": 101,
    "name": "Alice",
    "role": "Admin"
}

# 2. Key to check
key_to_find = "email"

# 3. Use the 'in' keyword to check existence
if key_to_find in user_data:
    print(f"Key '{key_to_find}' exists. Value: {user_data[key_to_find]}")
else:
    print(f"Key '{key_to_find}' does not exist in the dictionary.")
```

```python
#10. Calculate Average Marks

# 1. Input list of marks
marks = [45, 78, 90, 33, 60]

# 2. Calculate the sum of all marks
total_sum = sum(marks)

# 3. Get the total number of students (count)
count = len(marks)

# 4. Calculate the average
average = total_sum / count

# 5. Display the results
print(f"Marks List: {marks}")
print(f"Total Sum: {total_sum}")
print(f"Number of Students: {count}")
print(f"Average Mark: {average}")
```
